```python
theta = 3
n = 100
sample = uniform.rvs(-theta, 2*theta, size=n)
plt.hist(sample, density=True, bins=20)
plt.title("Гистограмма выборки из $U[{}, {}]$".format(-theta, theta))
pass
```

Вместо работы с $X_1, \ldots, X_n \sim \mathcal{U}[-\theta, \theta]$ рассмотрим задачу $X_1, \ldots, X_n \sim \mathcal{U}[0, \theta]$. Переход между такими задачами это просто сдвиг на $\theta$ вправо и $\dfrac{\theta}{2}$ влево соответственно.
В случае равномерного распределения $X_1, \ldots, X_n \sim \mathcal{U}[0, \theta]$ для нахождения $\hat{\theta} = \underset{\theta \in \Theta}{argmax} f_\theta\lr{X_1, \ldots, X_n}$ продифференцировать  функцию правдоподобия не получится, ведь функция плотности разрывна
$$
f(x) =
      \begin{cases}
        \dfrac{1}{\theta},\hspace{0.5cm} x \in [0, \theta] ~\\
        ~\\
        0, \hspace{0.7cm} x \in (-\infty, 0) \cup (\theta, +\infty)
      \end{cases}
$$

Функция правдоподобия имеет вид $f_\theta\lr{X_1, \ldots, X_n} = \lr{\dfrac{1}{\theta}I\lr{0 \le X_1, \ldots, X_n \le \theta}}^n = \dfrac{1}{\theta^n}I\lr{0 \le X_1, \ldots, X_n \le \theta}$

Там, где $f_\theta$ не равна нулю, она равна некоторой константе $\dfrac{1}{\theta^n}$ , которую надо максимизировать, то есть надо минимизировать $\theta$. Но сделать её меньше $X_{(n)}$ не получится, так как иначе не выполнится условие под индикатором. Следовательно, $\boxed{\widehat{\theta}_n = X_{(n)}}$ будет искомой ОМП.
Получим функцию распределения статистики $\widehat{\theta}_n = X_{(n)}$ :

$$P_{\theta}(X_i \le t) = \frac{t}{\theta}, \hspace{1cm} t \in [0, \theta]$$
$$P_{\theta}(X_{(n)} \le t) = P_{\theta}(X_1 \le t, \ldots, X_n \le t) = P_{\theta}(X_1 \le t)\cdot \ldots \cdot P_{\theta}(X_n \le t) = \lr{\frac{t}{\theta}}^n$$

Проверим состоятельность нашей статистики, то есть покажем, что $\forall \theta \in \Theta \quad \widehat{\theta}_n \overset{P_{\theta}}{\longrightarrow} \theta$

Для $0 < \eps < \theta$ имеем
$$P_\theta\lr{\abs{X_{(n)} - \theta} > \eps} = P_\theta\lr{X_{(n)} > \theta + \eps} + P_\theta\lr{X_{(n)} < \theta - \eps} = 0 + \lr{\dfrac{\theta - \eps}{\theta}}^n \to 0$$
$\qed$

# Задача №3

image.png

$$X = \lr{X_1, X_2, \ldots} \sim Beta(\alpha, \beta)$$
$$\E{X} = \dfrac{\alpha}{\alpha + \beta}, \qquad \V{X} = \dfrac{\alpha\beta}{(\alpha+\beta)^2(\alpha + \beta + 1)}$$

$$T_n = \dfrac{\sum_{i=0}^{n-1} X_i - n\E{X_1}}{\sqrt{n\V{X_1}}}$$
Из ЦПТ
$$T_n \overset{d}{\underset{n\rightarrow\infty}{\longrightarrow}} \mathcal{N}(0,1)$$

```python
alpha = 7
beta = 3
n = 1000000

E = alpha / (alpha + beta)
V = alpha * beta / ((alpha + beta + 1) * (alpha + beta)**2)

sample = sps.beta.rvs(alpha, beta, size=n)
plt.hist(sample, density=True, bins=50, label='гистограмма выборки')
plt.plot(np.linspace(0, 1, 1000), sps.beta.pdf(np.linspace(0, 1, 1000), alpha, beta), label='плотность Beta({},{})'.format(alpha, beta))
plt.legend()
pass
```

```python
def T_stat(m) :
    return (np.sum(sample[0:m]) - m*E) / np.sqrt(m*V)

T = np.zeros(n+1)
for i in range(1, n) :
    T[i] = (T[i-1] * np.sqrt((i-1)*V) + sample[i] - E) / np.sqrt(i*V)

plt.plot(T)
plt.title('Статистика $T_n$')
plt.xlabel('n')
plt.ylabel('$T_n$')
pass
```

Сходимости $T_n$ не наблюдается как минимум потому, что из ЦПТ мы знаем, что
$$T_n \overset{d}{\underset{n\rightarrow\infty}{\longrightarrow}} \mathcal{N}(0,1)$$,
 то есть единственным возможным пределом последовательности статистик мог бы служить нуль, но по графику для большой выборки видно обратное

```python
plt.hist(x=T, density=True, bins=50, label='гистограмма статистики $T_n$')
plt.plot(np.linspace(-3, 3, 1000), sps.norm.pdf(np.linspace(-3, 3, 1000)), label='плотность $\mathcal{N}$(0,1)')
plt.title('Проверка ЦПТ на примере $T_n$ статистики')
plt.legend()
pass
```

Из графика заключаем, что с ростом размера выборки выполнение ЦПТ всё более заметно.

# Задача №4

image.png


$$X = \lr{X_1, \ldots, X_{100}} \sim Pareto(\alpha, 1), \quad \alpha \in (0, 5)$$
$$F_\alpha(t) = 1 - \dfrac{1}{t^\alpha}, \quad t \ge 1$$
Заметим, что в силу симметрии относительно ординаты $\dfrac{1}{2}$ имеем $1 - F_\alpha(t)$ распределена так же как и $F_\alpha(t)$ на $[0,1]$. В самом деле,
$$1 - \dfrac{1}{t^\alpha} = \dfrac{1}{2} + \lr{\dfrac{1}{2} - \dfrac{1}{t^\alpha}}, \qquad \dfrac{1}{t^\alpha} = \dfrac{1}{2} - \lr{\dfrac{1}{2} - \dfrac{1}{t^\alpha}}$$
Поэтому статистика
$$G(X, \alpha) = -\sum_{i=1}^n \log\lr{1-F_\alpha (X_i)} = \alpha \sum_{i=1}^n \log(X_i)$$
является центральной и распределена как $\Gamma(n,1)$. Откуда

$$\gamma = P_\alpha\lr{u_\frac{1-\gamma}{2} < G(X, \alpha) < u_\frac{1+\gamma}{2}} = P_\alpha\lr{u_\frac{1-\gamma}{2} < \alpha \sum_{i=1}^n \log(X_i) < u_\frac{1+\gamma}{2}} = P_\alpha\lr{\dfrac{u_\frac{1-\gamma}{2}}{\sum_{i=1}^n \log(X_i)} < \alpha < \dfrac{u_\frac{1+\gamma}{2}}{\sum_{i=1}^n \log(X_i)}}$$

где $u_p$ есть $p-$квантиль распределения $\Gamma(n,1)$.

```python
import warnings
warnings.filterwarnings("ignore")

precision = 0.95
n = 100
grid = 1 + np.arange(n)


def theta_hat(sample, theta):
    return theta * np.sum(np.log(sample))


def interval(sample, precision):
    lower = sps.gamma(a=n, scale=1).ppf((1 - precision)/2) / np.log(np.prod(sample))
    upper = sps.gamma(a=n, scale=1).ppf((1 + precision)/2) / np.log(np.prod(sample))
    return float(lower), float(upper)


plt.figure(figsize=(20, 10))
for i in range(1, 50, 7) :
    sample = sps.pareto.rvs(size=n, b=i/100)
    intervals = np.array([interval(sample[:n], precision) for n in grid])
    plt.fill_between(grid, intervals.T[0], intervals.T[1], alpha=0.5, label='$\\alpha={}$'.format(i/100))
    plt.title('Доверительный интервал параметра для $Pareto(\\alpha)$ для уровня доверия {}'.format(precision))
    plt.xlabel('размер выборки n')
    plt.ylabel('оценка $\widehat{\\alpha}$')
    plt.ylim(0, 20)
    plt.xlim(0, 40)
    plt.legend()
pass
```

# Задача №5

image.png

```python
alpha = 5
beta = 1
n = 100

def F(x, alpha, beta): # Weibull_cdf
    return 1 - np.exp(-(beta*x)**alpha)
```

Лемма. $$F_\xi(\xi) \sim \mathcal{U}[0,1]$$
Д-во:

cdf $F_\xi$ монотонно неубывает, значит
$$P\lr{F_\xi(\xi) \le F_\xi(x)} = P\lr{\xi \le x} = F-xi(x) \in [0,1]$$
Откуда
$$P\lr{F_\xi(\xi) \le y} = y \in [0,1] \implies F_\xi(\xi) \sim U[0,1]$$
$\blacksquare$

Итак,
$$\xi \sim F_\xi^{-1}\lr{\mathcal{U}[0,1]}$$
В нашем случае
$$F_\xi(x) = 1 - exp\lr{-(\beta x)^\alpha}, \quad x > 0,\ \alpha > 0,\ \beta > 0$$
То есть
$$F^{-1}_\xi(t) = \dfrac{\sqrt[\alpha]{-\log(1 - t)}}{\beta}$$

```python
def f(t, alpha, beta): # F^{-1}
    return np.power(-np.log(1-t), 1/alpha) / beta


sample = f(np.random.rand(n), alpha, beta)
plt.hist(sample, density=True, bins=50)
plt.title("Гистограмма выборки из Weibull({}, {})".format(alpha, beta))
pass
```

```python
X = sample.copy()
X.sort()
X
```

Метод спейсингов:
$$D_i(\theta) = F_\theta(X_{i+1}) - F_\theta(X_{i}),\quad i=0,\ldots,n, \quad X_{0} := -\infty, X_{n+1} := +\infty$$
$$\hat{\theta} = arg\max_\theta{\sum_{i=0}^n \log D_i(\theta)}$$

```python
def spacing_method(params):
    a = params[0]
    b = params[1]
    D = np.zeros(n+1)
    D[0] = F(X[0], a, b) - 0
    D[n] = 1 - F(X[n-1], a, b)
    for i in range(1, n):
        D[i] = F(X[i], a, b) - F(X[i-1], a, b)
    return -np.sum(np.log(D))
```

```python
from scipy.optimize import minimize
res = minimize(spacing_method, x0=[3,1])
res
```

Наблюдаем достаточно точный ответ. Метод хорошо показал себя применительно к данной задаче.
